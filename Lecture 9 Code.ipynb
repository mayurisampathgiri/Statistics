```python
#Lecture 9 Code
#Simple linear regression
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy
data = pd.read_csv("C:\\Users\\Mayur\\OneDrive\\Documents\\PythonStat\\OxygenPurity.csv")
x = data['x'].values
y = data['y'].values
slope, intercept, r, pvalue, std_err = stats.linregress(x, y)
print(stats.linregress(x, y))
def retFunc(x):
    return slope * x + intercept
model = list(map(retFunc, x))
plt.scatter(x, y)
plt.plot(x, model)
plt.show()
```
